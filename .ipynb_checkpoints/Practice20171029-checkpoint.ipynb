```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
with open ('appledaily.csv','w', encoding='utf-8-sig') as f:
    for rec in soup.select('.even a'):
        title    = rec.select('h1')[0].text
        category = rec.select('h2')[0].text
        time     = rec.select('time')[0].text
        link     = rec['href']
        #print (category, time, time, link)
        f.write('{}\t{}\t{}\t{}\n'.format(category, time, title, link))
        
```

```python
news=[]
for rec in soup.select('.even a'):
    new_dic = {}
    new_dic['title'] = rec.select('h1')[0].text
    new_dic['category'] = rec.select('h2')[0].text
    new_dic['time'] = rec.select('time')[0].text
    new_dic['link'] = rec['href']
    news.append(new_dic)
    
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def pageDetail(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    title = soup.select('h1')[0].text
    content = soup.select('.ndArticle_margin p')[0].text
    dt = soup.select('.ndArticle_creat')[0].text.replace('建立時間：','')
    clicked = None
    try:
        clicked = soup.select('.ndArticle_view')[0].text
    except:
        pass
    return {'title':title, 'content':content, 'dt':dt, 'clicked':clicked}   
```

```python
pageDetail('https://tw.news.appledaily.com/life/realtime/20171029/1230695/「會賺錢vs不會賺錢」/u3000這五大特質要看清楚')
```

```python
domain = 'http://www.appledaily.com.tw'
url = 'https://tw.appledaily.com/new/realtime/{}'

news_ary=[]
for i in range(1,11):
    res = requests.get(url.format(i))
    soup = BeautifulSoup(res.text, 'html.parser')
    
    for rec in soup.select('.rtddt a'):
        #print(rec)
        news_dic = {}
        news_dic['link'] = rec['href']
        news_dic['category'] = rec.select('h2')[0].text
        content_dic =pageDetail(news_dic['link'])
        news_dic.update(content_dic)
        news_ary.append(news_dic)
```

```python
import requests
import pandas

headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36 OPR/48.0.2685.52'
}

url  = 'https://rent.591.com.tw/home/search/rsList?is_new_list=1&type=1&kind=1&searchtype=1&region=1&section=5&firstRow={}&totalRows=749'
rent_591= []
for i in range(0,25):
    res = requests.get(url.format(i*30), headers = headers)
    jd = res.json()
    df = pandas.DataFrame(jd['data']['data'])
    rent_591.append(df)
```

```python
#print (rent_591)
housedf =pandas.concat(rent_591)
print (housedf)
```

```python
housedf = housedf[['id','address','floor','allfloor','layout','area','price','browsenum_all']]
print (housedf)
```

```python
housedf['price'] = housedf['price'].map(lambda e: int(e.replace(',','')))
```
