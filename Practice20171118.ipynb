```python
import pandas

df = pandas.read_csv('https://raw.githubusercontent.com/ywchiu/tibamepy/master/data/salary.csv')
df
```

```python
%pylab inline
df.plot(kind ='scatter', x='year', y='salary')
```

```python
from matplotlib import pyplot as plt
X= df[['year']]
Y= df[['salary']].values
plt.xlabel('year')
plt.ylabel('salary')
plt.scatter(X, Y, color='black')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X, Y)
```

```python
print('Intercept', regr.intercept_) # regr.intercept_ 當無年資時新水多少
print('coefficient', regr.coef_) # regr.coef_ 每增加一年薪水增加多少
```

```python
regr.predict(X) # 根據年資產生的預測結果
```

```python
X= df[['year']]
Y= df[['salary']].values
plt.xlabel('year')
plt.ylabel('salary')
plt.scatter(X, Y, color='black')
plt.plot(X, regr.predict(X), color='blue')
plt.show()
```

```python
salaries = pandas.DataFrame([{'year': 6}, {'year':8}])
print (regr.predict(salaries))
```

## 簡單線性回歸

```python
from sklearn.externals import joblib # google search 'sklearn preserve model'
joblib.dump(regr, 'regression.pkl') # *.pkl file is python's API object
```

```python
clf = joblib.load('regression.pkl')
salaries = pandas.DataFrame([{'year': 6}, {'year':8}])
print (clf.predict(salaries))
```

## 多項式線性回歸

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_reg = PolynomialFeatures(degree=2)
X_ = poly_reg.fit_transform(X)
regr = LinearRegression()
regr.fit(X_, Y)
```

```python
plt.scatter(X, Y)
plt.plot(X_, regr.predict(X_), color='blue', linewidth = 3)
plt.show()
```

```python
X2 = X.sort_values('year')
X2_ = poly_reg.fit_transform(X2)
plt.scatter(X, Y)
plt.plot(X2, regr.predict(X2_), color='blue', linewidth = 3)
plt.show()
```

## 多元回歸模型

```python
import pandas
df  = pandas.read_csv('https://raw.githubusercontent.com/ywchiu/tibamepy/master/data/house-prices.csv')
df.head()
```

```python
dummy_brick = pandas.get_dummies(df['Brick']) # Brick => Yes & No
dummy_neighbor = pandas.get_dummies(df['Neighborhood']) # Neighborhood => East, West & North
house = pandas.concat([df, dummy_brick, dummy_neighbor], axis=1)
house
```

```python
del house['No']
del house['West']
del house['Brick']
del house['Neighborhood']
del house['Home']
house.head()
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
X = house[['SqFt', 'Bedrooms', 'Bathrooms', 'Offers', 'Yes', 'East', 'North']]
Y = house['Price'].values
regr.fit(X, Y)
```

```python
#regr.predict(X)
print (regr.intercept_)
print (regr.coef_)
```

```python
house_price = pandas.DataFrame([{
        'SqFt': 2000,
        'Bedrooms':4,
        'Bathrooms': 2,
        'Offers': 1,
        'Yes': 1,
        'East': 1,
        'North':0
    }])
print (regr.predict(house_price))
```

## 評估迴歸模型

```python
import statsmodels.api as sm

X2 = sm.add_constant(X) #加上常數
est = sm.OLS(Y, X2) # Y:預測變數(應變數)，X2:自變數
est2 = est.fit()
print (est2.summary())
```
